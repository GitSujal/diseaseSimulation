# Importing Modules into the Program

```python
from people import People
import random
import numpy as np
import matplotlib.pyplot as plt
```

### Initialising some variables if the program has to run without any arguments.

```python
population = []
INIT_POP = 2000
INIT_INFECTED = 300
NUM_COLS = 40
NUM_ROWS = 40
NUM_STEPS = 10
infection_probability = INIT_INFECTED/INIT_POP
```

## Initialising population with people. Population is a list of people objects. 
1. Each people have their own random location given as Xposition and Yposition.
2. Each people have a certain probability of initial infection calculated from initial infected population.
3. Each people have a certain random resistance ranging from (0,1).
4. Each people have a infection time which is zero for non infected person and it increases by one for each timestep for infected person.
5. Each infected person is plotted as Red dot on the plot and each unifected person is plotted as blue on the plot.

```python
numberOfInfected = 0;
infected_cells = []
total_infected = []
for i in range(0,INIT_POP):
    person=People(NUM_ROWS,NUM_COLS,infection = np.random.choice([0,1],p=[1-infection_probability,infection_probability]) ,infection_time=0)
    #person.print_out()
    if person.infected:
        infected_cells.append((person.xpos,person.ypos))
        numberOfInfected += 1;
    population.append(person)
    person.plot_people()
plt.show()
print(numberOfInfected)
total_infected.append(numberOfInfected)
```

## Following activity occurs in each timestep of the simulation
1. Each person move randomly either one x postion, y position or both based on neighbourhood selection ie Von Numann or Moore Neighbour.
2. If an uninfected person is in a cell with an infected person there is a chance of infection. If the chance of infection is greater than the person's resistance the person gets infected.
3. The grid is plotted for each timestep and the output figure is saved.

```python
for timestep in range(NUM_STEPS):
    print("\n###################### TIMESTEP", timestep, "#####################\n")
    print("Total Infected ", numberOfInfected)
    total_infected.append(numberOfInfected)
    for person in population:
        person.move_people(NUM_ROWS,NUM_COLS)
        numberOfInfected = person.check_infection(infected_cells,numberOfInfected)
        person.plot_people()
    filename="./Output/Outputfile"+str(timestep)+".png"
    plt.savefig(filename)
    plt.show()
    plt.clf()
    plt.cla()
    plt.close()
    print("File saved as ",filename)
    
```

# Plotting the growth of infected Population over time
1. This plot shows how the infected population increases over time. Usually this curve resembles an exponential growth.

```python
plt.plot(range(0,NUM_STEPS+1),total_infected)
plt.savefig("growth.png")
plt.ylabel("Infected Population")
plt.xlabel("TIMESTEP")
plt.title("Growth of infected Population over time")
plt.show()
```
